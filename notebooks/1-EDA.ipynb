```python
# Import library
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading dataset
data1 = pd.read_csv("../data/db1.csv")

# Checking instances/labels and imbalance label
data1.shape
data1.info()
data1['Diabetes_binary'].value_counts(normalize=True) # => Imbalance data with ~84% of class 0
                                                      # Bias increase

# Visualize the imbalance of class 0 and 1
sns.countplot(x='Diabetes_binary', data= data1)
plt.xlabel('Diabetes')
plt.ylabel('Instances')
```
